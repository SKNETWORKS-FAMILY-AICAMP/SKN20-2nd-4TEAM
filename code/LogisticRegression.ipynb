```python
X = df.drop(columns='Target',axis=1)
y = df.Target

from sklearn.preprocessing import StandardScaler  # <- 이게 맞음
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline


pipe = Pipeline([
    ("scale", StandardScaler()),
    ("kbest", SelectKBest(score_func=f_classif)),   # 어느 컬럼이 제일 영향력이 짱 쎌까
    ("clf", LogisticRegression(max_iter=1000))
])

param = {
    "kbest__k": [5, 10, 15, 20, "all"]  # 몇 개까지의 컬럼을 학습 시키는게 좋은가?
}

grid = GridSearchCV(pipe, param, cv=5, scoring="roc_auc")
grid.fit(X, y)
```

```python
grid.best_score_      
grid.best_params_     
```

```python
best_model = grid.best_estimator_
# 예: X_test가 있다면
# from sklearn.metrics import roc_auc_score
# roc_auc_score(y_test, best_model.predict_proba(X_test)[:,1])
```

```python
kbest = best_model.named_steps["kbest"]
selected_cols = X.columns[kbest.get_support()]
list(selected_cols)
```

```python
from sklearn.model_selection import train_test_split


df2=df.copy()
df_data = df2[['Debtor','Tuition fees up to date',
'Curricular units 1st sem (credited)',
'Curricular units 1st sem (enrolled)',
'Curricular units 1st sem (evaluations)',            
'Curricular units 1st sem (approved)',
'Curricular units 1st sem (grade)'   ,
'Curricular units 1st sem (without evaluations)',    
'Curricular units 2nd sem (credited)',
'Curricular units 2nd sem (enrolled)',
'Curricular units 2nd sem (evaluations)',            
'Curricular units 2nd sem (approved)',
'Curricular units 2nd sem (grade)'   ,
'Curricular units 2nd sem (without evaluations)','Target']]

X=df_data.drop(columns='Target',axis=1)
y=df_data.Target

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=45,stratify=y)
model = LogisticRegression(max_iter=1000)

model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred, digits=4))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 0) 최종 df에서 라벨 확정만 사용 (필요 시)
df_hm = df_data.copy()
# 재학생이 섞였으면 주석 해제
# df_hm = df_hm[df_hm['Target'].isin(['Dropout','Graduate'])].copy()

# 1) Target → 숫자
label_map = {'Dropout': 1, 'Graduate': 0}
df_hm['Target_num'] = df_hm['Target'].map(label_map)

# 2) 상관계수 (Spearman)
#    문자열 Target은 빼고 숫자들만 계산
corr = df_hm.drop(columns=['Target']).corr(method='spearman', numeric_only=True)

# 3) (옵션) Target과의 |상관| 큰 순서로 행/열 재정렬
order = corr['Target_num'].abs().sort_values(ascending=False).index
corr = corr.loc[order, order]

# 4) Target_num을 우측 하단으로 확실히 보내고 싶으면(이미 그럴 확률 높지만)
cols = [c for c in corr.columns if c != 'Target_num'] + ['Target_num']
corr = corr.loc[cols, cols]

# 5) 히트맵
plt.figure(figsize=(18, 18))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=True,
            cbar_kws={'shrink': .8})
plt.title("Spearman Correlation (including Target)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1) 재학생 제외
df_bin = df[df['Target'].isin(['Dropout', 'Graduate'])].copy()

# 2) Target → 숫자(히트맵에 포함시키려면 필요)
label_map = {'Dropout': 1, 'Graduate': 0}
df_bin['Target_num'] = df_bin['Target'].map(label_map)

# 3) Spearman 상관계수 (수치형만)
corr = df_bin.drop(columns=['Target']).corr(method='spearman', numeric_only=True)

# 4) (옵션) Target과의 |상관| 기준으로 정렬 + Target_num을 맨 끝으로
order = corr['Target_num'].abs().sort_values(ascending=False).index
corr = corr.loc[order, order]
cols = [c for c in corr.columns if c != 'Target_num'] + ['Target_num']
corr = corr.loc[cols, cols]

# 5) 히트맵
plt.figure(figsize=(18, 18))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={'shrink': .8})
plt.title("Spearman Correlation (Dropout vs Graduate only)")
plt.tight_layout()
plt.show()
```
